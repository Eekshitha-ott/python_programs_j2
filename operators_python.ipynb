# operators in python

## 6 types of operators in python
1. arithematic operators
2. assignment operators
3. comparision operators
4. logical operators
5. membership operators
6. identity operators

```python
# arithematic operators - +,-,*,/,//,%
a=10
b=20
c=20
d=30
print(a+b)
print(a-c)
print(b*d)
print(d/a)
print(d//a)
print(d%a)
```

```python
#assignment operators - +=,-=,/=,//=,%=
a=100
a+=50
print(a)
```

```python
b=200
b-=400
print(b)
```

```python
c=50
c*=100
print(c)
```

```python
#comparision operators - >,<,>=,<=,==,!=
a=10
b=20
c=20
d=30
print(a>b)
print(b==c)
print(b<c)
print(b!=c)
print(b>=a)
print(b<=a)
```

```python
#logical operator- and,or,not
"""
and - both conditions are true thrn results is True , rest case it false
or - either one condition is true the result is true , rest case result is false
and truth table
T T = T
T F = F
F T = T
F F = F
or truth table
T T = T
T F = T
F T = T
F F = F
"""

a=10
b=20
c=20
d=30
print(a>b and b==c)
print(a<b and c==b)
print(a>b or b==c)
print(not a>b and b==c)
```

```python
# membership operater - in,not in 
# validating members are  present in that sequence or group
names=["eekshitha","sanvika","keren","yashasree"]
print("eekshitha" in names)
print("sanvika" not in names)
print("Eekshitha" not in names)
print("nitya" in names)
```

```python
#identity operator - is,is not
a=10000
b=10000
print(a is b)
print(id(a))
print(id(b))
```

```python
a=10
b=10
print(a is b)
print(id(a))
print(id(b))
```
